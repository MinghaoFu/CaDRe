```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data
npz = np.load('${PROJECT_ROOT}/workspace/climate-project/General/data/nonparametric_bt/data.npz')
zt = npz['zt']

# Reshape and process zt
zt = zt.reshape(160000, 3, 2)

# Create a figure
plt.figure(figsize=(10, 8))

# Plot for the first type of zt
plt.scatter(zt[:, 0, 0], zt[:, 0, 1], alpha=0.5, color='blue', label='zt-1')

# Plot for the second type of zt
plt.scatter(zt[:, 1, 0], zt[:, 1, 1], alpha=0.5, color='green', label='zt')

# Plot for the third type of zt
plt.scatter(zt[:, 2, 0], zt[:, 2, 1], alpha=0.5, color='red', label='zt+1')

# Add labels and title
plt.xlabel('Coordinate X')
plt.ylabel('Coordinate Y')
plt.title('Scatter Plot of Different Types of zt')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
