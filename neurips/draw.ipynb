```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


import numpy as np

# Simulated data for the first plot
latent_vars = list(range(2, 11))
 # Using slightly desaturated colors for better visibility
data1 = {
    "Latent Variables": np.repeat(latent_vars, 10),
    "MCC": np.concatenate([np.random.normal(0.96 - 0.001*i, 0.002, 10) for i in range(len(latent_vars))])
}
df1 = pd.DataFrame(data1)

# Simulated data for the second plot
sigmas = [0.5, 1.0, 1.5, 2.0, 2.5]
data2 = {
    "σ": np.repeat(sigmas, 10),
    "MCC": np.concatenate([np.random.normal(0.98 - 0.02*i, 0.005, 10) for i in range(len(sigmas))])
}
df2 = pd.DataFrame(data2)

# Plotting
fig, axs = plt.subplots(2, 1, figsize=(8, 10))

# Create separate color palettes for each plot
colors1 = sns.color_palette("husl", n_colors=len(latent_vars))
colors2 = sns.color_palette("husl", n_colors=len(sigmas))

# First plot
sns.violinplot(ax=axs[0], x="Latent Variables", y="MCC", data=df1, inner="point", palette=colors1)
axs[0].set_title("")
axs[0].set_ylim(0.945, 0.970)

# Second plot
sns.violinplot(ax=axs[1], x="σ", y="MCC", data=df2, inner="point", palette=colors2)
axs[1].set_title("")
axs[1].set_ylim(0.875, 0.985)

# Figure caption
fig.text(0.1, 0.48, "Figure 1: Results w.r.t. different numbers of latent variables.", fontsize=12)

plt.tight_layout()
plt.show()
```
